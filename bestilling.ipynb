```python
class AnnoCorpus:
    def __init__(self, step=10, limit=30000, **kwargs):
        keywords = kwargs
        keywords['limit'] = limit
        corpus = dict()
        from_year = keywords['from_year']
        to_year = keywords['to_year']
        for year in range(from_year,to_year, step):
            keywords['from_year'] = year
            keywords['to_year'] = year + step
            corpus[year] = dh.Corpus(**keywords).frame[['dhlabid','urn', 'title', 'year', 'ddc', 'genres', 'literaryform','city']]
        self.corpus = corpus

    def corpus_num(self, sequence):
        if sequence in self.corpus.keys():
            res = self.corpus[sequence]
        elif sequence in range(len(self.corpus.keys())):
            res = self.corpus[list(self.corpus.keys())[sequence]]
        else:
            res = pd.DataFrame()
        return res
        
```

```python
a = AnnoCorpus(doctype='digibok', from_year=1945, to_year=2023, step=10, subject="kultur*")
```
