## Bestilling fra november 2023

Bestilling til DH-lab

Korpus - Barne- og ungdomslitteratur utgitt i Norge Tidsperiode: 1945 til i dag, med mulighet for fortløpende oppdatering Materialtype: Alle slags bøker og tegneserier og tidsskrift som er gitt ut for målgruppen barn og unge Dette inkluderer blant annet pekebøker, bildebøker, kapittelbøker, barneromaner, ungdomsromaner, lettlestbøker, seriebøker, diktsamlinger, dramatikk, faktabøker, faglitteratur, essays, biografier, leksikon mm Alt som er gitt ut for barn og unge MEN IKKE skolebøker Språk: Norsk bokmål, nynorsk, samiske språk, kvensk, skogfinsk, romani, romanes

Korpus - Dokumenter fra Kulturdepartementet, inkl i samarbeid med andre departement

Kulturdepartementet har hatt ulike navn: Kirke- og kulturdepartementet, Kultur- og likestillingsdepartementet, Kulturdepartementet, Kirke- og undervisningsdepartementet, Kyrkje- og undervisningsdepartementet, Kyrkjedepartementet, Kyrkje- og kulturdepartementet

Tidsperiode: 1945 til i dag, med mulighet for fortløpende oppdatering Materialtype: Rapporter, planer og strategier, NOU’er, Stortingsmeldinger, Prop’er

Alle temaer Språk: Norsk bokmål, nynorsk

Korpusene må gjerne inndeles i 10-år da vi ønsker å kunne foreta sammenligninger for å spore utvikling fra tiår til tiår

1945 – 1949 1950 – 1959

1960 – 1969

1970 – 1979

1980 – 1989

1990 – 1999

2000 – 2009

2010 – 2019

2020 – til i dag, med mulighet for fortløpende oppdatering

Vi ber om at korpusene utformes slik at vi kan foreta følgende operasjoner

Søke på ord-nivå, ordstamme Frekvenstelling, få satt opp frekvenslister for de aktuelle søkeordene

Undersøke frekvens av sammenstillinger (at noen ord vi definerer opptrer i nærheten av hverandre, men også åpne søk/få frem hvorvidt det er sammenstillinger vi ikke har tenkt på) Kollokasjoner, se hvilke ord de vi søker på opptrer sammen med

Konkordans, slik at vi kan lese teksten rundt søkeordet, noen setninger før og etter

Metadata, slik at vi bare ved et tastetrykk kan få tilgang til hele det aktuelle dokumentet (få korpuset på excel?)

Det må også være mulig å søke på bilder. Barne- og ungdomslitteraturen er til en stor grad visuell. Kan det som Jørgen Schyberg holder på med; Maken hjelpe oss der? Politiske ord som medvirkning, mangfold eller fornorsking vil jo gjerne vises i bilder like mye som tekst.
